# PCA

```python
# Applying PCA Fit Transform to dataset
pca =PCA(n_components=13)# there are 13 columns
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1, color= 'green')
```

```python
#Final dataframe
final_wine = pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_wine
```

```python
final_wine.Type.value_counts()
```

```python
#Plotting in scatter plot
plt.figure(figsize=(15,10))
sns.scatterplot(data =final_wine)
```

# Hierarchical clustering

```python
#dendogram visualization
dendogram = sch.dendrogram(sch.linkage(wine_normal,'complete'))
```

```python
#Create Hierarchical cluster
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
#Creating dataframe
y = pd.DataFrame(hc.fit_predict(wine_normal),columns=['HC_clusters'])
y
```

```python
y['HC_clusters'].value_counts()
```

```python
hc.labels_
```

```python
# adding the predicted dataframe into dataset
wine1 = wine.copy()
wine1['HC_clusters']=hc.labels_
wine1
```

# K-Means

```python
# within-cluster sum-of-squares criterion 
wcss = []
for i in range(1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_normal)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('ELBOW GRAPH')
plt.xlabel('No.of clusters')
plt.ylabel('WCSS')
```

```python
#Build Cluster algorithm using K=3
clusters=KMeans(3,random_state=30).fit(wine_normal)
clusters
```

```python
clusters.labels_
```

```python
#Creating dataframe
wine2 = wine.copy()
wine2['KMean_clusters'] = clusters.labels_
wine2
```

```python
wine2['KMean_clusters'].value_counts()
```
